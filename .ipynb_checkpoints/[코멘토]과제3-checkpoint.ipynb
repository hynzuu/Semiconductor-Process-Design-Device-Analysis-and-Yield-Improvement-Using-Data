## MOL 공정 분석 report: Silicide & contact CD data trend 분석

```python
df['Step_Description'].unique()
```

```python
silicide_df=df[df['Step_Description']=='Silicide 두께 ']
silicide_df
```

###  Lot별로 확인

```python
df2=silicide_df.pivot_table(index='Lot ID',values=['Target','Spec high','Spec low','Value(Å)'],aggfunc='mean').reset_index()
df2
```

```python
plt.figure(figsize=(6,5))
sns.lineplot(data=df2, x='Lot ID', y='Target', label='Target')
sns.lineplot(data=df2, x='Lot ID', y='Spec high', label='Spec high')
sns.lineplot(data=df2, x='Lot ID', y='Spec low', label='Spec low')
sns.lineplot(data=df2, x='Lot ID', y='Value(Å)', label='Value(Å)')
plt.legend()
plt.show()
```

```python
# Target보다  낮은 경향 발견
# 두께 타겟팅을 위해 올려야
```

### Lot별 , 웨이퍼 위치별로 확인

```python
# Lot ID, Wafer ID, CHIP_X_Y별
df4=silicide_df.pivot_table(index=['Lot ID','Wafer ID','CHIP_X_Y'],values=['Target','Spec high','Spec low','Value(Å)'],aggfunc='mean').reset_index()
df4
```

```python
df4['new_ID'] = df4['Lot ID'].astype(str) +str('_') +df4['Wafer ID'].astype(str) +str('_')+df4['CHIP_X_Y'].astype(str)
df4
```

```python
plt.figure(figsize=(9,5)  )
sns.lineplot(data=df4, x='new_ID', y='Target', label='Target')
sns.lineplot(data=df4, x='new_ID', y='Spec high', label='Spec high')
sns.lineplot(data=df4, x='new_ID', y='Spec low', label='Spec low')
sns.lineplot(data=df4, x='new_ID', y='Value(Å)', label='Value(Å)')
plt.legend()
plt.show()
```

```python
# 산포를 줄여야 함
# spec low 보다 낮은 값들이 많은 것을 확인 할 수 있음
# 한선에 걸리는 데이터 확인 필요
# spec low 와 spec high수준으로 진행된 chip 간의 소자나 수율 특성 차이 여부 확인 필요
```

### Silicide thickness vs wafer 중심으로부터의 거리

```python
plt.figure(figsize=(10, 6))
for wafer_id in silicide_df['Wafer ID'].unique():
    wafer_data = silicide_df[silicide_df['Wafer ID'] == wafer_id]
    plt.plot(wafer_data['CHIP_X_Y'], wafer_data['Value(Å)'], marker='o', linestyle='-', label=wafer_id)

plt.title('Silicide Thickness vs Wafer Position')
plt.xlabel('Wafer Position (CHIP_X_Y)')
plt.ylabel('Silicide Thickness (Å)')
plt.legend()
plt.grid(True)
plt.show()
```
