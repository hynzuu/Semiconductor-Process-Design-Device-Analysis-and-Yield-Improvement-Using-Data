## 2) 각 Lot 마다의 평균 Gate ACI CD 를 value, spec high, target, spec low 와 함께 plotting 해보세요. (분산형 or 꺾은선형) (* 평균값 처리
작업 효율화 위해서는 피벗테이블 활용 필요)

```python
df['Lot ID'].unique()
```

```python
filter_df=df[["Stepseq", "Step_Description", "Tkin_time", "Lot ID", "Wafer ID", "Item_id", "subitem_id", "Value(Å)", "Spec high", "Target", "Spec low"] ]
```

```python
lot_avg=filter_df.groupby('Lot ID')['Value(Å)'].mean().reset_index()
lot_avg
```

```python
# 기준값 추가
lot_avg['Spec_high'] = filter_df['Spec high'].iloc[0]
lot_avg['Target'] = filter_df['Target'].iloc[0]
lot_avg['Spec_low'] = filter_df['Spec low'].iloc[0]
lot_avg
```

```python
# 데이터 시각화
plt.figure(figsize=(10, 6))
plt.plot(lot_avg['Lot ID'], lot_avg['Value(Å)'], label='Average Value', marker='o')
plt.plot(lot_avg['Lot ID'], lot_avg['Spec_high'], label='Spec High', linestyle='--')
plt.plot(lot_avg['Lot ID'], lot_avg['Target'], label='Target', linestyle='--')
plt.plot(lot_avg['Lot ID'], lot_avg['Spec_low'], label='Spec Low', linestyle='--')
plt.xlabel('Lot ID')
plt.ylabel('Average CD1 ACI CD')
plt.title('Average Gate ACI CD per Lot')

plt.xticks(rotation=45)
plt.legend()
plt.show()
```

## 3) Equipment & Chamber data 를 묶어 표현한 후 설비&챔버 별로 Gate ACI CD 를 plotting 해보세요. (상자 수염 box plot 형)

```python
df.columns
```

```python
filter_df=df[["Equipment ID","Chaber ID","Stepseq", "Step_Description", "Tkin_time", "Lot ID", "Wafer ID", "Item_id", "subitem_id", "Value(Å)", "Spec high", "Target", "Spec low"] ]
```

```python
filter_df['Equip_Chamber'] = filter_df['Equipment ID'] + '_' + filter_df['Chaber ID'].astype(str)
```

```python
filter_df
```

```python
plt.figure(figsize=(12, 6))
filter_df.boxplot(column='Value(Å)', by='Equip_Chamber')
plt.title('Gate ACI CD by Equipment and Chamber')
plt.suptitle('')
plt.xlabel('Equipment and Chamber')
plt.ylabel('CD1 ACI CD')
plt.xticks(rotation=45)
plt.show()
```

그래프에 나타난 데이터의 경향성과 특성을 분석해 보겠습니다.

### 1. 첫 번째 그래프: Gate Etch 1st Step CD1 ACI CD

- **데이터 개요**: 이 그래프는 Gate Etch 첫 번째 단계에서 CD1 ACI CD 값을 시간에 따라 보여줍니다.
- **경향**:
  - 값이 목표선 주위에서 변동하지만 일반적으로 사양 한계 내에 유지됩니다.
  - 간헐적으로 사양 상한선을 초과하거나 사양 하한선 아래로 내려가는 경우가 있습니다.
- **관찰**:
  - 전반적으로 공정이 제어되고 있지만 약간의 변동이 있습니다.
  - 이상치에 대해 원인을 파악하여 주의를 기울여야 합니다.

### 2. 두 번째 그래프: Gate Etch 2nd Step CD1 ACI CD

- **데이터 개요**: 이 그래프는 Gate Etch 두 번째 단계에서 CD1 ACI CD 값을 시간에 따라 보여줍니다.
- **경향**:
  - 데이터 포인트에 뚜렷한 상승 추세가 있어 시간이 지남에 따라 CD 값이 증가하고 있습니다.
  - 값이 변동하지만 사양 상한선에 가까워지고 있습니다.
- **관찰**:
  - 상승 추세를 조사하여 사양 상한선을 초과하지 않도록 해야 합니다.
  - 목표 범위 내에 값을 유지하기 위해 정기적인 모니터링과 조정이 필요합니다.

### 3. 세 번째 그래프: Lot당 평균 Gate ACI CD

- **데이터 개요**: 이 그래프는 Lot당 평균 CD1 ACI CD 값을 보여줍니다.
- **경향**:
  - 평균 값이 대부분 목표 범위 내에 있지만 변동이 있습니다.
  - 몇 개의 Lot에서 값이 사양 하한선을 초과하는 경우가 있습니다.
- **관찰**:
  - 평균 값은 일반적으로 사양을 충족하지만, 변동을 해결해야 합니다.
  - 일관성을 유지하기 위해 지속적인 모니터링과 품질 관리가 필요합니다.

### 요약 및 권장 사항

- **전반적인 공정 제어**: 공정은 전반적으로 제어되고 있지만 일부 변동성과 경향성이 주목할 만합니다.
- **집중 영역**:
  - **첫 번째 단계의 이상치**: 간헐적으로 발생하는 스파이크와 딥의 원인을 조사해야 합니다.
  - **두 번째 단계의 상승 추세**: 상승 추세를 제어하고 값을 사양 내에 유지하기 위한 조치를 취해야 합니다.
  - **평균 값의 일관성**: Lot당 평균 값의 변동을 최소화하기 위해 지속적인 공정 제어가 필요합니다.
